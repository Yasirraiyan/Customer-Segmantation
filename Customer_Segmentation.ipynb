```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# ডেটাসেট তৈরি করা
data = {
    'Age': [23, 45, 31, 35, 43, 25, 21, 33, 37, 41],
    'Income': [35000, 60000, 47000, 52000, 70000, 28000, 26000, 42000, 54000, 65000]
}
df = pd.DataFrame(data)

# K-means ক্লাস্টারিং প্রয়োগ করা
kmeans = KMeans(n_clusters=2)
kmeans.fit(df)

# ক্লাস্টার লেবেল যোগ করা
df['Cluster'] = kmeans.labels_

# সেন্ট্রয়েড পয়েন্ট বের করা
centroids = kmeans.cluster_centers_

# ক্লাস্টার প্লট করা
plt.scatter(df['Age'], df['Income'], c=df['Cluster'], cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', marker='X')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Customer Segmentation with K-means Clustering')
plt.show()
```
